```python
import numpy as np
import pandas as pd
```

```python
total_student_df = pd.read_csv("Resources/students_complete.csv")
total_school_df = pd.read_csv("Resources/schools_complete.csv")
```

```python
total_school_df
```

```python
#district metrics
school_total = total_school_df['School ID'].count()
student_total = total_student_df['Student ID'].count()
buget_total = total_school_df['budget'].sum()
math_avr = total_student_df['math_score'].mean()
reading_avr = total_student_df['reading_score'].mean()

#passing math
num = total_student_df[total_student_df.math_score >= 70].count()
num = num / student_total * 100
per_passing_math = num['Student ID']

#passing reading
num1 = total_student_df[total_student_df.reading_score >= 70].count()
num1 = num1 / student_total * 100
per_passing_reading = num1['Student ID']


passing_rate = (math_avr + reading_avr)/2

```

```python

columns = ["Total Schools", "Total Students", "Budget Total", "Math Average %", "Reading Average %", "Students Passing Math %", "Students Passing Reading %", "Total Students Passing %" ]
df_ = pd.DataFrame(columns=columns)


df_.at[0,'Total Schools'] = school_total
df_.at[0,'Total Students'] = student_total
df_.at[0,'Budget Total'] = buget_total
df_.at[0,'Math Average %'] = math_avr
df_.at[0,'Reading Average %'] = reading_avr
df_.at[0,'Students Passing Math %'] = per_passing_math
df_.at[0,'Students Passing Reading %'] = per_passing_reading
df_.at[0,'Total Students Passing %'] = passing_rate

#district metric dataframe
df_
```
